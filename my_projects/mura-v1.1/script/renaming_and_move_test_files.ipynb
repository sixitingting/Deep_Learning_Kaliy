```python
def move_files(name): 
    PATH = (f'{NB_DIR}/').replace('script/','')+f'data/MURA-v1.1/valid/{name}/'
    folders = os.listdir(PATH) #alle pasientene innefor et felt 
    for file in folders:
        print('\n')
        patient_folder = f'{PATH}{file}/' 
        patient_studies = os.listdir(patient_folder) #alle pasientene 
    
        for study in patient_studies:  #alle studiene til en pasient (1-2 studier)
            study_folder = f'{patient_folder}{study}/'
            study_images = os.listdir(study_folder)
        
            for image in study_images: 
                var = name+'_'+file+'_'+study+'_'+ image #patient12104_study1_image3.png
                print(var)
                old_file = os.path.join(study_folder, image)
                new_file = os.path.join('/home/sathiesh/Deep_Learning_Kaliy/my_projects/mura-v1.1/data/test', var)
                os.rename(old_file, new_file)
```

```python
for i in XR:
    move_files(i);
```
